```python
from moviepy.editor import VideoFileClip, AudioFileClip

# 1. โหลดวิดีโอและเช็คว่าโหลดได้จริง
video = VideoFileClip("Video1.mp4")
if video.duration is None:
    print("Error: ไม่สามารถอ่านค่าความยาววิดีโอได้")
else:
    # 2. โหลดเสียง
    new_audio = AudioFileClip("audio_ori.mp3")
    
    # 3. ตัดเสียง (เช็คให้ชัวร์ว่าวิดีโอไม่ยาวกว่าไฟล์เสียงต้นฉบับ)
    duration = min(video.duration, new_audio.duration)
    new_audio = new_audio.subclip(0, duration)

    # 4. ใส่เสียงใหม่
    final = video.set_audio(new_audio)

    # 5. Export
    final.write_videofile(
        "Vdo_new_audio.mp4",
        fps=video.fps,
        codec="libx264",
        audio_codec="aac"
    )

    # ปิด resource
    video.close()
    new_audio.close()
    final.close()
```

```python
from moviepy.editor import VideoFileClip, AudioFileClip

# โหลดวิดีโอ
video = VideoFileClip("Video1.mp4")

# โหลดเสียงใหม่ และตัดให้ยาวเท่าวิดีโอ
new_audio = AudioFileClip("audio_ori.mp3").subclip(0, video.duration)

# ใส่เสียงใหม่แทน
final = video.set_audio(new_audio)

# export
final.write_videofile(
    "Vdo_new_audio.mp4",
    fps=video.fps,
    codec="libx264",
    audio_codec="aac",
    audio=True
)

# ปิด resource
video.close()
new_audio.close()
final.close()
```

```python
from moviepy.editor import VideoFileClip, AudioFileClip

#โหลด Vdo
video = VideoFileClip('Video1.mp4')

#โหลดเสียงใหม่เข้ามา พร้อมทำให้ความยาวเสียงใหม่ยาวเท่าวิดีโอ
new_audio = AudioFileClip("audio_ori.mp3").subclip(0, video.duration)

'''ใช้เวลานานกว่า คลิปจากมือถือจะเพี้ยน แต่คลิปความเร็วปกติ'''
#แทนเสียงใหม่ใน Vdo
final = video.set_audio(new_audio)

# export เป็นไฟล์จริง
final.write_videofile(
    'Vdo_new_audio.mp4',
    fps=video.fps , # ปรับ fps ให้ตรงกับคลิป
    codec='libx264', # ตัวเข้ารหัสเสียง (Audio Codec)
    audio_codec='aac', # มาตรฐานเสียงของ mp4
    audio=True # บังคับว่าต้องมีเสียง
)

#ปิด Vdo, audio
video.close()
new_audio.close()
final.close()
```
