```python
from moviepy import VideoFileClip

#โหลด Vdo
video =  VideoFileClip('Video1.mp4')

#แยกไฟล์เสียงจาก Vdo เป็น mp3
video.audio.write_audiofile('audio_ori.mp3')

#ปิด Vdo
video.close()
```

```python
from moviepy import VideoFileClip, AudioFileClip

#โหลด Vdo
video = VideoFileClip('Video1.mp4')

#โหลดเสียงใหม่เข้ามา พร้อมทำให้ความยาวเสียงใหม่ยาวเท่าวิดีโอ
new_audio = AudioFileClip("audio_ori.mp3").subclipped(0, video.duration)

#แทนเสียงใหม่ใน Vdo
final = video.with_audio(new_audio)

# export เป็นไฟล์จริง
final.write_videofile(
    'Vdo_new_audio.mp4',
    fps=video.fps if video.fps else 30, # ป้องกันกรณีค่า fps เป็น None
    codec='libx264',
    audio_codec='aac',
    preset='ultrafast' # ช่วยให้การประมวลผลเฟรมไม่เพี้ยนจากการบีบอัดมากไป
)

#ปิด Vdo, audio
video.close()
new_audio.close()
final.close()
```
