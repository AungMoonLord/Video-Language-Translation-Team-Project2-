```python
from moviepy import VideoFileClip, AudioFileClip

#โหลด Vdo
video = VideoFileClip('Video1.mp4')

#โหลดเสียงใหม่เข้ามา พร้อมทำให้ความยาวเสียงใหม่ยาวเท่าวิดีโอ
new_audio = AudioFileClip("audio_ori.mp3").subclipped(0, video.duration)

'''ใช้เวลาเร็วกว่า คลิปจากมือถือไม่เพี้ยน แต่แก้ความเร็วคลิปยาก'''

#แทนเสียงใหม่ใน Vdo
final = video.with_audio(new_audio)

# export เป็นไฟล์จริง
final.write_videofile(
    'Vdo_new_audio.mp4',
    fps=video.fps , # ปรับ fps ให้ตรงกับคลิป
    codec='libx264', # ตัวเข้ารหัสเสียง (Audio Codec)
    audio_codec='aac', # มาตรฐานเสียงของ mp4
    audio=True # บังคับว่าต้องมีเสียง ป้องกันการลบเสียงเองเพราะคิดว่า error
)

#ปิด Vdo, audio
video.close()
new_audio.close()
final.close()
```

```python
from moviepy.editor import VideoFileClip, AudioFileClip

#โหลด Vdo
video = VideoFileClip('Video1.mp4')

#โหลดเสียงใหม่เข้ามา พร้อมทำให้ความยาวเสียงใหม่ยาวเท่าวิดีโอ
new_audio = AudioFileClip("audio_ori.mp3").subclip(0, video.duration)

'''ใช้เวลานานกว่า คลิปจากมือถือจะเพี้ยน แต่คลิปความเร็วปกติ'''
#แทนเสียงใหม่ใน Vdo
final = video.set_audio(new_audio)

# export เป็นไฟล์จริง
final.write_videofile(
    'Vdo_new_audio.mp4',
    fps=video.fps , # ปรับ fps ให้ตรงกับคลิป
    codec='libx264', # ตัวเข้ารหัสเสียง (Audio Codec)
    audio_codec='aac', # มาตรฐานเสียงของ mp4
    audio=True # บังคับว่าต้องมีเสียง ป้องกันการลบเสียงเองเพราะคิดว่า error
)

#ปิด Vdo, audio
video.close()
new_audio.close()
final.close()
```
