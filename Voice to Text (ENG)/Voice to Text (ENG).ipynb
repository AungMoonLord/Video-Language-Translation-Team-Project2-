```python
import warnings
from transformers import pipeline
import torch
```

```python
# voice1

# 1. ปิดการแสดงผลข้อความเตือน (Warning)
warnings.filterwarnings("ignore")

# ตั้งค่า Device
device = "cuda:0" if torch.cuda.is_available() else "cpu"
torch_dtype = torch.float16 if torch.cuda.is_available() else torch.float32

# โหลดโมเดล
pipe = pipeline(
    "automatic-speech-recognition",
    model="openai/whisper-large-v3",
    torch_dtype=torch_dtype,
    device=device,
)

audio_file_path = "voice1.mp3"

# 2. เปลี่ยน language เป็น "thai" หรือลบออกเพื่อให้มัน Auto-detect
result1 = pipe(
    audio_file_path,
    generate_kwargs={"language": "english"} # เปลี่ยนตรงนี้เป็น thai
)

print("-" * 50)
print(result1["text"])
print("-" * 50)
```

```python
# voice1


# ตั้งค่า Device (ถ้ามีการ์ดจอ NVIDIA ให้ใช้ "cuda:0", ถ้าไม่มีใช้ "cpu")
device = "cuda:0" if torch.cuda.is_available() else "cpu"
torch_dtype = torch.float16 if torch.cuda.is_available() else torch.float32

print(f"Loading Model on: {device}...")

# สร้าง Pipeline
# นี่คือทางลัดที่ง่ายที่สุดในการเรียกใช้ model
pipe = pipeline(
    "automatic-speech-recognition",
    model="distil-whisper/distil-large-v3",
    torch_dtype=torch_dtype,
    device=device
)

# ใส่ไฟล์เสียง (เปลี่ยนชื่อไฟล์ตรงนี้เป็นไฟล์ของคุณ)
# รองรับ .mp3, .wav, .mp4 (บางกรณี)
audio_file = "voice1.mp3"

print("Transcribing...")

# สั่งถอดเสียง
# สำคัญมาก! ต้องใส่ return_timestamps=True เพื่อให้รู้เวลาเริ่ม-จบของประโยค
result_1 = pipe(
    audio_file,
    return_timestamps=True,
    generate_kwargs={"language": "english"} # บังคับว่าเป็นภาษาอังกฤษเพื่อความชัวร์
)

# 5. แสดงผลลัพธ์
print("\n--- Full Text ---")
print(result_1["text"])
```

```python
# voice1
import re
from funasr import AutoModel
import torch

# 1. ฟังก์ชันล้าง Tags (เหมือนเดิม)
def clean_sensevoice_text(text):
    clean_text = re.sub(r'<\|.*?\|>', '', text)
    return clean_text.strip()

# 2. โหลดโมเดล SenseVoice
print("⏳ Loading SenseVoice...")
model = AutoModel(
    model="iic/SenseVoiceSmall",
    trust_remote_code=True,
    device="cuda:0" if torch.cuda.is_available() else "cpu"
)

# 3. ระบุไฟล์เสียง
audio_file = "voice1.mp3" # <-- แก้ชื่อไฟล์ตรงนี้

# 4. เริ่มถอดเสียง
print(f"▶ Processing: {audio_file}")
res = model.generate(
    input=audio_file,
    cache={},
    language="auto",
    use_itn=True,
    batch_size_s=60,
)

# 5. ดึงข้อความและล้างให้สะอาด
raw_text = res[0]['text']
final_text = clean_sensevoice_text(raw_text)

print("\n--- Final English Text ---")
print(final_text)
```
