```python
from TTS.api import TTS

# โหลดโมเดล KhanomTan ภาษาไทย
tts = TTS(
    model_name="tts_models/th/khanomtan/v1.1",
    gpu=False   # Colab CPU ก็ใช้ได้
)

# ข้อความภาษาไทย
text = "อากาศสดชื่น ธรรมชาติอุดมสมบูรณ์ รุ่งรวยวัฒนธรรม ท่ามกลางโลกสมัยใหม่ที่พัฒนาอย่างรวดเร็ว"

# สร้างไฟล์เสียง
tts.tts_to_file(
    text=text,
    file_path="thai_voice.wav"
)

print("สำเร็จแล้ว สร้างไฟล์ thai_voice.wav")
```
