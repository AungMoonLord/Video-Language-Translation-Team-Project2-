```python
from gtts import gTTS
from IPython.display import Audio, display
```

```python
# --- 1. สร้างฟังก์ชัน (เครื่องมือผลิตเสียง) ---
def make_voice(text, filename):
    print(f"⏳ กำลังแปลง: {text[:40]}...")  # โชว์ข้อความช่วงต้น
    
    # แปลงเสียง
    tts = gTTS(text=text, lang='th', slow=False) 
    tts.save(filename)
    
    # แสดงผล
    print(f"✅ บันทึกเสร็จ: {filename}")
    display(Audio(filename, autoplay=False))
    print("-" * 30) # ขีดเส้นคั่นสวยๆ
```

```python
long_text = """
ซีรีส์แนวกฎหมาย-คอมเมดี้ ชีวิตมนุษย์ที่จะพาผู้ชมติดตามเรื่องราวการเดินทางสุดวุ่นวาย
ของผู้พิพากษาอินฟลูเอนเซอร์ชื่อดังผู้หมกมุ่นในความสำเร็จและวัตถุนิยม 
จนกระทั่งวันหนึ่งได้พบกับเหตุการณ์สุดช็อกบางอย่าง
"""
make_voice(long_text, "series_test.mp3")
```
